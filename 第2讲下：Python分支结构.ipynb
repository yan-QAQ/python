```python
import random 
# 1. 随机产生两个个位数
number1 = random.randint(0, 9)
number2 = random.randint(0, 9)

# 2. 如果number1 < number2, 交换它们的值
if number1 < number2:
    number1, number2 = number2, number1

# 3. 提示用户回答问题"请问 number1 - number2 = ?"
answer = eval(input("请问 " + str(number1) + " - " + str(number2) + " = ? "))

# 4. 检查答案是否正确并显示结果
if number1 - number2 == answer:
    print("回答正确")
else:
    print("回答错误")
```

### 案例3-3 计算身体质量指数BMI


```python
# 提示用户输入体重(kg)、身高(m)
weight = eval(input("请输入体重（千克）: "))
height = eval(input("请输入身高（米）: "))

# 计算 BMI
bmi = round(weight / (height * height), 2)

# 展示计算结果
print("BMI的值是" + str(bmi))
if bmi < 18.5:
    print("超轻")
elif bmi < 25:
    print("标准")
elif bmi < 30:
    print("超重")
else:
    print("肥胖")
```

### 案例3-4 判断闰年
#### 编写程序，提示用户输入年份，然后判断是否是闰年，如果是，则输出“闰年”，否则输出“平年”。


### 案例3-4-1 判断闰年 (逻辑表达式)


```python
year = eval(input("请输入年份: "))
# 使用逻辑表达式判断是否为闰年 
if (year % 4 == 0 and year % 100 != 0) or    (year % 400 == 0):
    print("闰年")
else:
    print("平年")
```

### 案例3-4-2 判断闰年 (嵌套if语句)

```python
year = eval(input("请输入年份: "))

# 使用嵌套if语句判断是否为闰年
if year % 4 != 0:
    print("平年")
else:
    if year % 100 != 0:
        print("闰年")
    else:
        if year % 400 == 0:
            print("闰年")
        else:
            print("平年")
```
