```python
x = 5
print("before: x is", x)

def print_info():
    print("************")
    print("hello world!")
    print("************")

print_info()
x = x + 2
print_info()
 
print("after: x is", x)
```

### 函数调用栈

```python
def max(a, b):
    if a > b:
        result = a
    else:
        result = b
    return result

def main():
    i = 2
    j = 5
    k = max(i, j)
    print(k)

main()
```

（1）当程序调用一个函数时，程序控制权就会转移到被调用函数上
（2）当函数执行结束时，被调用函数就会将程序控制权交还给调用者


### 案例5-1 显示素数（使用函数改进）

```python
## 案例 4-7
from math import sqrt

count = 0  # 记录已找到的素数个数
number = 2  # 将要被测试的自然数
    
# 重复寻找素数
while count < 50:
    is_prime = True  # 标志变量，当前判断的数是否为素数

    # 判断自然数是否为素数
    for divisor in range(2, int(sqrt(number)) + 1): 
        # 只要有一个divisor能整除number，number就不是素数
        if number % divisor == 0:
            is_prime = False          
            break

    # 打印素数，增加素数个数count
    if is_prime:
        count += 1

        # 打印素数，不换行，每个数字占5位空间，右对齐
        print(format(number, "5d"), end="")
        # 每打印10个素数，换行
        if count % 10 == 0:
            print()

    # number增1，继续判断下一个自然数是否为素数
    number += 1
```

```python
## 改进

from math import sqrt

# 判断自然数是否为素数
def is_prime(num):
    for divisor in range(2, int(sqrt(num)) + 1): 
        # 只要有一个divisor能整除num，num就不是素数
        if num % divisor == 0:
            return False          
    return True
                
count = 0  # 记录已找到的素数个数
number = 2  # 将要被测试的自然数
    
# 重复寻找素数
while count < 50:
    # 打印素数，增加素数个数count
    if is_prime(number):
        count += 1
        
        # 打印素数，不换行，每个数字占5位空间，右对齐
        print(format(number, "5d"), end="")
        # 每打印10个素数，换行
        if count % 10 == 0:
            print()

    # number增1，继续判断下一个自然数是否为素数
    number += 1

```

2. 编写测试程序，显示从2010年到2020年每年的天数

```python
def number_of_days(year):
    if is_leap_year(year):
      return 366
    else:
      return 365

def is_leap_year(year):
    return year % 400 == 0 or (year % 4 == 0 and year % 100 != 0)

for year in range(2000, 2010 + 1):
    print(year, number_of_days(year))
```

```python
def print_month(year, month):
    print_month_title(year, month)    
    print_month_body(year, month)
  
def print_month_title(year, month): 
    print("       ", get_month_name(month), " ", year)
    print("-----------------------------")
    print(" Sun Mon Tue Wed Thu Fri Sat")
 
def print_month_body(year, month): 
    
    start_day = get_start_day(year, month)

    numberofdays_month = get_numberofdays_month(year, month)

    i = 0
    for i in range(start_day):
        print("    ", end = "")

    for i in range(1, numberofdays_month + 1):
        print(format(i, '4d'), end = "")

        if (i + start_day) % 7 == 0:
            print()

def get_month_name(month): 
    if month == 1:
        month_name = "January"
    elif month == 2:
        month_name = "February"
    elif month == 3:
        month_name = "March"
    elif month == 4:
        month_name = "April"
    elif month == 5:
        month_name = "May"
    elif month == 6:
        month_name = "June"
    elif month == 7:
        month_name = "July"
    elif month == 8: 
        month_name = "August"
    elif month == 9:
        month_name = "September"
    elif month == 10:
        month_name = "October"
    elif month == 11:
        month_name = "November"
    else:
        month_name = "December"

    return month_name

# 返回某月第一天的星期数 
def get_start_day(year, month): 
    START_DAY_FOR_JAN_1_1800 = 3

    # 从1/1/1800到month/1/year的总天数
    total_number_of_days = get_total_number_of_days(year, month)

    return (total_number_of_days + START_DAY_FOR_JAN_1_1800) % 7

# 从1/1/1800到month/1/year的总天数
def get_total_number_of_days(year, month): 
    total = 0

    # 从1800到year的天数
    for i in range(1800, year):
        if is_leap_year(i):
            total = total + 366
        else:
            total = total + 365

    # 从year/1/1到month/1的天数
    for i in range(1, month):
        total = total + get_numberofdays_month(year, i)

    return total

def get_numberofdays_month(year, month): 
    if (month == 1 or month == 3 or month == 5 or month == 7 or
        month == 8 or month == 10 or month == 12):
        return 31

    if month == 4 or month == 6 or month == 9 or month == 11:
        return 30

    if month == 2:
        return 29 if is_leap_year(year) else 28

    return 0

def is_leap_year(year): 
    return year % 400 == 0 or (year % 4 == 0 and year % 100 != 0)

 
year = eval(input("Enter full year (e.g., 2001): "))
month = eval(input(("Enter month as number between 1 and 12: ")))

print_month(year, month)
 
```

### 5.1 求n（n是正整数）的阶乘

```python
def get_factorial(number):
    result =1
    while number>0:
        result *= number
        number -=1
    return result

print("factorial 6=", get_factorial(6))
print("factorial 3=", get_factorial(3))
print("factorial 100=", get_factorial(100))
```

### 5.4 求从1到n 的平方和

```python
def sum_of_square(n):
    result = 0
    for number in range(1,n+1):
        result += number*number
    return result

print("sum of square 3:", sum_of_square(3))
print("sum of square 6:", sum_of_square(6))
print("sum of square 10:", sum_of_square(10))
```
