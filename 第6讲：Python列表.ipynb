## Python列表

类似字符串，列表也是由一系列值构成的序列。字符串中值是字符，而列表中值可以是任何类型

列表常量由中括号[]包围，其中元素由逗号,隔开

列表元素可以是任何类型的Python对象，甚至可以是另一个列表

列表可以为空，空列表由空的中括号[]表示

```python
print([1, 24, 76])
```

```python
print(['red', 'yellow', 'blue'])
```

```python
print(['red', 24, 98.6])
```

```python
print([1, [5, 6], 7])
```

```python
print([])
```

## 遍历列表

```python
# 这种遍历方法只能读取列表中的每个元素，不能修改
numbers = [1, 3, 5, 7, 9] 
for num in numbers:
    print(num)
    
```

```python
#这种遍历方法使用range()和len()函数生成列表的全部索引组成的序列，遍历过程中通过索引就可以修改元素的值
numbers = [1, 3, 5, 7, 9]
for i in range(len(numbers)):
    numbers[i] = numbers[i] + 1
print(numbers)
```

## 内置函数和列表

Python中有很多内置函数可以接收列表作为参数，直接处理列表，不用自己编写循环

```python
nums = [3, 41, 12, 9, 74, 15]
print(len(nums))
print(max(nums))
print(min(nums))
print(sum(nums))
print(sum(nums)/len(nums))
```

```python
total = 0
count = 0
while True :
    inp = input('Enter a number: ')
    if inp == 'done' : break
    value = float(inp)
    total = total + value     
    count = count + 1

average = total / count
print('Average:', average)
```

```python
numlist = list()
while True :
    inp = input('Enter a number: ')
    if inp == 'done' : break
    value = float(inp)
    numlist.append(value)

average = sum(numlist) / len(numlist)
print('Average:', average)
```

### 案例9-1 列表求和

```python
def sum_of_list(param_list):
    total = 0
    for item in param_list:
        total += item
    return total

list1 = [1,2,3,4]
list2 = [3,5,8,2]
print(f"sum of {list1}",sum_of_list(list1))
print(f"sum of {list2}",sum_of_list(list2))

#print(f"sum of {list1}",sum(list1))
#print(f"sum of {list2}",sum(list2))

```

### 案例9-2 打印一个范围内的偶数

```python
def get_even_number(begin, end):
    result = []
    for item in range(begin, end):
        if item % 2 == 0:
            result.append(item)
        #return result
    return result

begin = 4
end = 15
print(f"begin = {begin}, end = {end},even number:", get_even_number(begin, end))

#data = [item for item in range(begin, end) if item %2 ==0]
#print(f"begin = {begin}, end = {end},even number:", data)
```

###  案例9-3 删除列表中的元素

```python
def remove_elements_form_list(lista,listb):
    for item in listb:
        lista.remove(item)
    return lista


lista = [3,5,7,11,13]
listb = [7,11]
print(f"from {lista} remove {listb}, result:", remove_elements_form_list(lista,listb))

lista = [3,5,7,11,13]
listb = [7,11]
data = [item for item in lista if item not in listb]
print(f"from {lista} remove {listb}, result:", data)
```

### 案例9-4 获取列表中单个元素

```python
def get_unique_list(lista):
    result = []
    for item in lista:
        if item not in result:
            result.append(item)
    return result

lista = [10,20,30,10,20]
print(f"source list {lista}, unique list:",get_unique_list(lista))


#print(f"source list {lista}, unique list:",list(set(lista)))
```
