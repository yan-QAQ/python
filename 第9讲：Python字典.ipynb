## Python字典的创建

字典是一系列键和值的映射关系，每个键仅能映射到一个值。键和值的组合叫做键值对，构成字典的元素

使用内置函数dict()可以创建一个空字典，常量{}也表示一个空字典

字典中的键值对使用冒号:隔开，每个键值对之间使用逗号,隔开

键必须是不可变类型的对象，比如，字符串，数字，元组等

```python
d1 = dict()
d1
```

```python
d2 = {}
d2
```

```python
d3 = {'dog':'狗', 'cat':'猫'}
d3
```

## 访问字典中的元素

类似序列类型，也可以使用中括号运算符[]访问字典中的单个元素，索引值(或者叫键)可以是几乎任何类型

```python
d3 = {'dog':'狗', 'cat':'猫', 'panda':'熊猫'}
>>> d3['dog']
```

```python
d3['tiger']
```

字典对象还提供了一个get()方法也可以访问字典中的元素，并且访问不存在的键时不会像[]那样报错

```python
d3 = {'dog':'狗', 'cat':'猫', 'panda':'熊猫'}
```

```python
d3['tiger']
```

```python
d3.get('dog')
```

```python
print(d3.get('tiger'))
```

```python
d3.get('tiger', 'undefined')
```

## 字典的长度

内置函数len()可以返回任何容器对象的元素个数，对于字典来说，len()返回字典中键值对的个数

```python
d3 = {'dog':'狗', 'cat':'猫', 'panda':'熊猫'} 
len(d3)
```

## in和not in运算符

in 和 not in 也适用于字典，它们用于判断一个对象是否是字典中的键

```python
d3 = {'dog':'狗', 'cat':'猫', 'panda':'熊猫'} 
'dog' in d3
```

```python
'猫' in d3
```

为了判断一个对象是否是字典中的值，使用字典提供的values()方法，该方法返回字典中的所有值

```python
d3 = {'dog':'狗', 'cat':'猫', 'panda':'熊猫'} 
'猫' in d3.values()
```

```python
d3.values()
```

```python
list(d3.values())
```

## 字典元素的添加和更新

使用[]给字典的键赋值的方法添加和更新字典的元素

如果键不存在，就创建一个新的键值对，否则更新键对应的值

字典中键一旦存在就不能更改，只能更改值


```python
d1 = dict()
d1
```

```python
d1['dog'] = '狗'
d1['cat'] = '毛'
d1
```

```python
d1['cat'] = '猫'
d1
```

使用del可以删除字典中的键值对

使用字典的pop()方法也可以删除字典中的键值对

对于del和pop()来说，如果要删除的键不存在，都会报KeyError错误

```python
d1 = {'dog': '狗', 'cat': '猫'}
del d1['dog']
d1
```

```python
d1 = {'dog': '狗', 'cat': '猫'} 
d1.pop('dog')
d1
```

字典的clear()方法用于清空字典

使用del也可以直接删除字典对象

```python
d1 = {'dog': '狗', 'cat': '猫'}
d1.clear()
d1
```

```python
del d1
d1
```

## 字典的遍历

for循环也可以用于遍历字典，默认遍历字典的所有键

```python
ages = {'kim': 15, 'betty': 9, 'lisa': 28}

for key in ages:
    print(key, ages[key])     #字典中的键值对是无序的
```

### 案例11-1 学生成绩排序

```python

##1
students =[
    {"sno":101,"sname":"小张","sgrade":88},
    {"sno":102,"sname":"小王","sgrade":99},
    {"sno":103,"sname":"小李","sgrade":77},
    {"sno":104,"sname":"小赵","sgrade":66}
]
students_sort =sorted(students,key =lambda x:x["sgrade"])

print(f"source{students} sort result: {students_sort}")
```

```python
##2

students =[
    {"sno":101,"sname":"小张","sgrade":88},
    {"sno":102,"sname":"小王","sgrade":99},
    {"sno":103,"sname":"小李","sgrade":77},
    {"sno":104,"sname":"小赵","sgrade":66}
]
students_sort =sorted(students,key =lambda x:x["sgrade"],reverse=True)
print(students)
print(students_sort)
```
