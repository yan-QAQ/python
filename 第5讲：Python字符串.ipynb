## 遍历字符串

许多操作都需要以一次一个字符的方式处理字符串

具体来说，即从字符串开头开始，依次选取一个字符，为每个字符执行一些动作，一直持续直到字符串末尾，这种处理模式叫做遍历

```python
word = "python"
index = 0
while index < len(word): 
    letter = word[index]
    print(index, letter)
    index = index + 1
```

```python
word = "python"
for letter in word: 
    print(letter)
```

```python
word = "python"
for index in range(len(word)): 
    print(index, word[index])
```

## 搜索字符串


find函数是[]运算符的逆操作，返回参数字符letter在字符串word中的出现位置索引

如果字符letter没有在字符串word中出现，则函数返回-1

遍历序列，寻找目标元素，找到后返回其索引，这种处理模式叫做搜索

```python
def find(word, letter):
    index = 0
    while index < len(word):
        if word[index] == letter:
            return index
        index = index + 1
    return -1
print(find(word, "o"))
```

## 计数字符串

count函数计数字符letter在字符串word中出现的次数

这段程序演示了另一种处理模式，叫做计数器

```python
def count(word, letter):
    count = 0
    for ch in word:
        if letter == ch:
            count = count + 1
    return count
print(count("Python", "o"))
```

## 字符串切片

字符串的一个片段叫做切片，可以使用[start:end]运算符切取

[start:end]切取从索引start开始到end结束但不包含end的一个片段

如果end大于字符串长度，则切取到字符串末尾

```python
s = "Monty Python"
print(s[0:7])
print(s[6:11])
print(s[6:18])
print(s[5:1])
print(s[0:len(s)])
```

如果省略start，则默认从0开始

如果省略end，则默认到字符串末尾结束

```python
s = "Monty Python"
print(s[:5])
print(s[6:])
print(s[:])
```

切片高级语法: [start:end:step]，step表示切取片段时的步长，默认是1

start, end, step可以是负数，step为负数，表示反向切取

```python
s = "Monty Python"
print(s[0:5:2])
print(s[-6:12])
print(s[-1:-7:-1])
```

## 字符串格式化

### 第1种格式化字符串的方法：使用%表达式，%左侧是字符串模板，右侧是替换对象

```python
print("welcome to %s" % "python") 
```

```python
print("He is %d years old" % 30) 
```

```python
print("My name is %s and I am %d years old" % ("Jack", 30)) 
```

```python
print("interest is %-6.2f" % 1643434565.3678) 
```

```python
print("%o, %#o, %x, %#x" % (18, 18, 18, 18))
```

```python
print("%+6d, %07d" % (300, 300)) 
```

```python
print("%d%%" % 30)
```

 第2种格式化字符串的方法：使用str.format()方法，它增强了字符串格式化的功能。基本语法是通过 {} 和 : 来代替以前的 %

```python
"{} {}".format("hello", "world") # 不设置指定位置，按默认顺序
```

```python
"{0} {1}".format("hello", "world")  # 设置指定位置
```

```python
"{1} {0} {1}".format("hello", "world")  # 设置指定位置
```

```python
print("{:.2f}, {:+.2f}, {:+.2f}".format(3.1415, 3.1415, -3.1415)) 
```

```python
print("|{:>5d}|, |{:<5d}|, |{:^5d}|".format(8, 8, 8)) 
```

```python
print("|{:0>5d}|, |{:x<5d}|, |{:a^5d}|".format(8, 8, 8)) 
```

```python
print("{:,}, {:.2%}".format(1000000, 0.35))
```

```python
print("{:d}, {{0}}".format(200)) # 使用{}转义大括号
```
