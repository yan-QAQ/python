### 案例4-1 用while循环求1...10的和

```python
sum = 0
i = 1
while i <= 10:
    sum = sum + i
    print(i, sum)
    i = i + 1
print('while done')
print(i)
```

### 案例4-2 练习个位数减法（改进）

原案例：编写程序，随机生成两个个位数，向用户提出类似“8 – 3 = ?”这样的问题，检查用户回答是否正确并显示相应信息
改进：通过while循环，让用户一直输入答案直到答案正确为止

```python
import random

# 1. 随机产生两个个位数
number1 = random.randint(0, 9)
number2 = random.randint(0, 9)

# 2. 如果number1 < number2, 交换它们的值
if number1 < number2:
    number1, number2 = number2, number1

# 3. 提示用户回答问题"请问 number1 - number2 = ?"
answer = eval(input("请问 " + str(number1) + " - " + str(number2) + " = ? "))

# 4. 重复提问直到答案正确并显示结果
while number1 - number2 != answer:
    answer = eval(input("回答错误，请重试："))
print("回答正确")
```

### 案例4-3-0 求一组数的累加和

编写程序，读取并计算一组不确定个数的整数的累加和。输入0表明输入的结束。


```python
data = eval(input("请输入一个整数(0表明输入结束)："))

sum = 0
while data != 0:
    sum += data

    data = eval(input("请输入一个整数(0表明输入结束)："))

print("累加和为：", sum)
```

### 案例4-4：求整数序列的最大值


```python
largest_so_far = -1

print('Before', largest_so_far)

for num in [5, 9, 43, 21, 17] :
   if num > largest_so_far :
      largest_so_far = num
      
print('After', largest_so_far)
```

largest_so_far的初始值设为-1合理吗？

```python
# 改进
largest_so_far = None

print('Before', largest_so_far)

for num in [5, 9, 43, 21, 17] :
   if largest_so_far is None or num > largest_so_far :
      largest_so_far = num
      
print('After', largest_so_far)
```

### 案例4-5 计算数列a,aa,aaa,...的前n项和

```python
n = int(input("请输入整数n："))
a = int(input("请输入整数a："))

item = 0
sum = 0

for i in range(1, n + 1):
    item = item * 10 + a
    sum += item

print("累加和为：" + str(sum))
```

### 案例4-6 百钱百鸡


百钱百鸡是我国古代数学家张丘建在《算经》一书中提出的数学问题：鸡翁一值钱五，鸡母一值钱三，鸡雏三值钱一。百钱买百鸡，问鸡翁、鸡母、鸡雏各几何？
翻译成现代文是：公鸡5元一只，母鸡3元一只，小鸡1元三只，用100块钱买100只鸡，问公鸡、母鸡、小鸡各买多少只？
如果问题有解，输出所有解，每个解占一行，每个整数用空格隔开；如果问题无解，输出“无解”


```python
has_answer = False

for i in range(0, 21):
    for j in range(0, 34):
        k = 100 - i - j
        if 5 * i + 3 * j + k / 3 == 100:
            print(i, j, k)
            has_answer = True

if not has_answer:
    print("无解")
```

### 案例4-7 显示素数

显示自然数中的前50个素数，每行10个。素数即质数，是指除了1和它自身外，不能被其他自然数整除的数。

```python
# 该问题可以被分解为以下几个子任务：
# 判断一个数是否为素数
# 重复测试数字2, 3, 4, 5, ...是否为素数
# 统计已找到的素数个数
# 显示每个素数，每行显示10个
```

```python
from math import sqrt

count = 0  # 记录已找到的素数个数
number = 2  # 将要被测试的自然数
    
# 重复寻找素数
while count < 50:
    is_prime = True  # 标志变量，当前判断的数是否为素数

    # 判断自然数是否为素数
    for divisor in range(2, int(sqrt(number)) + 1): 
        # 只要有一个divisor能整除number，number就不是素数
        if number % divisor == 0:
            is_prime = False          
            break

    # 打印素数，增加素数个数count
    if is_prime:
        count += 1

        # 打印素数，不换行，每个数字占5位空间，右对齐
        print(format(number, "5d"), end="")
        # 每打印10个素数，换行
        if count % 10 == 0:
            print()

    # number增1，继续判断下一个自然数是否为素数
    number += 1
```
