## 3. 交换两个变量的值

#### 通用方法：引入一个中间变量


```python
a = 2
b = 3
temp = a
a = b
b = temp
print(a,b)
```

##### 使用Pythonic方法，同时赋值

```python
a = 2
b = 3
a, b = b, a
print(a, b)
```
